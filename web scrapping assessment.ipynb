# importing libraries

```python
import requests
from bs4 import BeautifulSoup
```

# scraping website

```python
url ='https://www.flipkart.com/mobiles/smartphones~type/pr?page=11&sid=tyy%2C4io'
```

# website response

```python
res = requests.get(url)
```

```python
res
```

# parsing

```python
soup = BeautifulSoup(res.text)
```

```python
soup
```

# mobile name

```python
mobile = soup.select('._4rR01T')
```

```python
mobile
```

```python
mobile_name=[]
```

```python
for item in mobile:
    mobile_name.append(item.text)
```

```python
mobile_name
```

# price

```python
price =soup.select('._1_WHN1')
```

```python
price
```

```python
price_value=[]
```

```python
for item in price:
    price_value.append(item.text)
```

```python
price_value
```

# mobile review

```python
review =soup.select("._13vcmD+ span")
```

```python
review
```

```python
mobile_review=[]
```

```python
for item in review:
    mobile_review.append(item.text)
```

```python
mobile_review
```

# people rating

```python
people =soup.select("._2_R_DZ span span:nth-child(1)")
```

```python
people
```

```python
people_rating=[]
```

```python
for item in people:
    people_rating.append(item.text)
```

```python
people_rating
```
